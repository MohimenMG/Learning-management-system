# Educational Management System Requirements
## Requirements List
1. Open institutions
2. Open classes
3. Create programs their courses, and their topics
4. Create exams and assessments for each course
5. Open departments
6. Register Instructors and assign them to their institution, department, and supervisor
7. Register Students and assign them to their program and institution
8. Assign a program to an institution
9. Assign a course to a program
10. Assign perquisites and semester to a course
11. Assign an instructor to a course's class room, time slot and set the class room max capacity 
12. Assign a student to a course's class room and time slot
13. Assign a student's course assessment score
14. Assign a student's course's final result
15. students can view their enrollment, assessment scores, exam submissions and courses results
16. instructors can create exams, assessments and set course and assessments results. they also can view their personal information and their students and class rooms

## Entity Analysis
### Main Entities
#### Institutions
1. Institutions Table
    - tracks all institutions/branches and their locations: `iti-smart village` is located at `Smart Villages Company, by Cairo-Alexandria Desert road` `giza` `egypt`
2. ClassRooms Table
    - tracks each unique class room in all locations and institutions: `iti-smart village` has a class room called `1B` 
3. Semesters Table
    - tracks the available semesters:  `Fall` `Spring` and `Summer`
4. time-slots Table
    - tracks available time slots: time-slot`1` starts at `08:00:00` and ends at `09:30:00` on `Sunday`
5. Programs Table
    - tracks programs details: the `mechanical engineering` program was created on `2015-01-01`
6. Departments Table
    - tracks the departments details: the `power electronics` department is available in many institutions
#### Programs
7. Courses Table
    - tracks courses basic info: the `thermodynamics` course was created on `2015-01-01` as part of the `mechanical engineering` program
8. Topics Table
    - tracks the topics chapters or lectures of a course: the `thermo-dynamics` course created on `2020-05-05` has `chapter 1: first law of thermodynamics` topic 
9. Exams Table
    - tracks all multiple choice exams: the `thermo-dynamics-1` course `thermodynamics-midterm-2020/05/05` exam was created on `2020-05-05`  
10. Questions Table
    - tracks the questions of a each exam and the topic related to each question: the `thermodynamics-midterm-2020/05/05` exam has `what is the first law of thermodynamics` question from `chapter 1: first law of thermodynamics` topic
11. Answers Table
    - tacks the answers choices of each question with a flag for the right answer:  `what is the first law of thermodynamics` question has an answer choice `energy can neither be created nor destroyed, only altered in form` which is `right`
12. Assessments Table
    - tracks every type of assessments that can be present in a course: courses can have assessments like `lecture presence`, `project-1` , `quiz-1`, `mid-term-exam`, `final-exam` etc
#### Instructors
13. Instructors Table
    - tracks instructors basic info: instructor `john smith` was born `1990-01-01` and his phone number is `010XXXXXX` his residence is `address string`
    - tracks hiring info: hired on `2023-09-10` with a basic salary of `5000`
    - department and supervisor info for example: `john smith`  works the `iti smart village` branch at the `cyber security` department and his supervisor is `joe adam` 
    - tracks instructors credentials: `email` and `password`
#### Students
14. Students Table
    - tracks students basic information for example:  student `john smith` was born `2000-01-01` and his phone number is `010XXXXXX` his residence is `address string`
    - tracks students enrollment info for example: enrolled in the `mechanical engineering` program on `2023-09-10` at `iti smart village` branch
    - tracks students login credentials like: `email` and `password`

### Relationship Entities
#### Institutions
1. InstitutionPrograms
    - tracks programs available for each institution: the `bio-medical engineering` program is available in `alexandria university` institution and manged by instructor `ahmed el ot`
2. InstitutionDepartments
    - tracks the departments in each institution: `alexandria university` has a `Nuclear Department` opened `2020-05-05`. and `ahmed` is the department manger who was assigned `2022-01-01`
#### Programs
3. ProgramCourses
    - tracks available courses for each programs:    
        - `thermodynamics-1` course is available to the `mechanical-engineering` program
    - some courses belongs to multiple programs: `math1` course belongs to all programs
4. CoursePrerequisites
    - tracks the prerequisites of each course:
        - `math3` has prerequisite `math1` and `math2` courses
        - `math1` course has `no prerequisite` courses
5. CourseSemester
    - tracks whether a course is available in the current semester: 
        - `thermodynamics-1` course is available during  the `fall` semester and is active
        - `thermodynamics-1` course is available during the `summer` semester but it is not `active` this semester
#### Instructors
6. ClassSchedule: 
    - tracks what courses and in which class rooms every instructor is teaching: 
        - `john smith` is teaching `math-1` course at class room `1B` which starts at `08:00:00` and ends at `09:30:00` and has a max capacity of `20` students
        - `john smith` is teaching `math-2` course at class room `2B` which starts at `08:00:00` and ends at `09:30:00` which has a max capacity of `20` students
#### Students
7. StudentEnrollments: 
    - tracks students enrolled courses' schedule and class rooms:
    `jack adam` is attending the `math-1` course at class rooms `1B` which starts at `08:00:00` and ends at `09:30:00`, taught by `john smith`. he enrolled on `2020-01-01 15:00:00`
    - the data timestamp can be used to analyze enrollment submissions patterns to aid in providing an optimized enrollment plan with high database server availability
8. AssessmentScores
    - tracks the score of every course assessment for every student: `john smith` achieved a score of `35` out of a total score of `40` in the `mid-term` assessment of the  `thermodynamics-1` course 
9. CourseResults
    - tracks the Status and results of a course: 
        - `john smith` enrolled in `thermodynamics-1` course on `2020-01-01`. the course status is `incomplete` and the course_result_percentage is `null`
        - `john smith` enrolled in `thermodynamics-1` course on `2020-01-01`. the course status is `pass` and the course_result_percentage is `86`
        - `john smith` enrolled in `thermodynamics-1` course on `2020-01-01`. the course status is `fail` and the course_result_percentage is `44.5`
10. ExamSubmissions 
    - tracks every student Answer for every exam: `john smith` submitted `energy can neither be created nor destroyed, only altered in form` Answer for the question `what is the first law of thermodynamics` on the `thermodynamics-midterm-2020/05/05` exam
    - the datetime is useful to gauge performance on time taken to answer a question

insert student with credentials proc
```sql
CREATE PROCEDURE InsertStudentWithLogin
    @StudentID INT,
    @FirstName VARCHAR(50),
    @LastName VARCHAR(50),
    @Gender CHAR(1),
    @DateOfBirth DATE,
    @EnrollmentDate DATE,
    @Phone VARCHAR(15),
    @Address VARCHAR(255),
    @City VARCHAR(50),
    @Country VARCHAR(50),
    @Email NVARCHAR(100),  -- Assuming university email as the username
    @Password NVARCHAR(50) -- Password for the login credential
AS
BEGIN
    BEGIN TRY
        -- Start a transaction
        BEGIN TRANSACTION;

        -- Insert student information into Students table
        INSERT INTO Students (student_id, student_first_name, student_last_name, student_gender,
            student_date_of_birth, student_institution_enrollment_date, student_phone, student_address,
            student_city, student_country)
        VALUES (@StudentID, @FirstName, @LastName, @Gender, @DateOfBirth, @EnrollmentDate,
            @Phone, @Address, @City, @Country);
        
        -- Validate password complexity rules
        IF NOT (
               PATINDEX('%[0-9]%', @Password) > 0 -- At least one number
           AND PATINDEX('%[A-Z]%', @Password) > 0 -- At least one uppercase letter
           AND PATINDEX('%[!@#$%^&*()]%', @Password) > 0 -- At least one symbol
        )
        BEGIN
            -- Password does not meet complexity rules
            THROW 51000, 'Password does must contain at least one number, one upper case letter and one symbol([!@#$%^&*()]).', 1;
        END

        -- Create a login for the student
        CREATE LOGIN [YourDatabaseName\@Email] WITH PASSWORD = @Password;

        -- Create a user for the student in the database
        USE YourDatabaseName;
        CREATE USER [YourDatabaseName\@Email] FOR LOGIN [YourDatabaseName\@Email];

        -- Add the student to the StudentRole
        ALTER ROLE StudentRole ADD MEMBER [YourDatabaseName\@Email];

        -- Commit the transaction
        COMMIT;
    END TRY
    BEGIN CATCH
        -- Rollback the transaction in case of an error
        IF @@TRANCOUNT > 0
            ROLLBACK;

        -- Handle the error (you may log or raise an error message)
        THROW;
    END CATCH;
END;

```
